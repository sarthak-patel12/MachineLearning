Before training as we can see that there is difference in the magnitude of the features of dataset
we will apply sacling

```python
from sklearn.preprocessing import StandardScaler

SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)
```

## Training and Testing

### Logistic Regression

As this is a multiclass classification we will use the first Logistic Regresion Algorithm

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression(verbose = 1)
lr.fit(X_train,y_train)
```

```python
y_test_predicted = lr.predict(X_test)
```

```python
y_test_predicted.reshape(-1,1).shape
```

```python
acrr = accuracy_score(y_test_predicted,y_test)
```

So we got the best accuracy using Logistic Regression
